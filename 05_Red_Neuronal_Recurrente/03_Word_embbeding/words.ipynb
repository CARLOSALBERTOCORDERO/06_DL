```python
def cos_similarity(A,B):
  mag_a = np.linalg.norm(A)
  mag_b = np.linalg.norm(B)
  sim = np.dot(A.T, B)/(mag_a*mag_b)
  return sim
```

```python
Betabel = np.array([1,0,0,0,0,0])
Naranja = np.array([0,0,1,0,0,0])
Mujer = np.array([0,1,0,0,0,0])

origin = [0],[0]
plt.quiver(*origin, Naranja,Betabel, color = ['r','g','b'], scale=5)
plt.show()
print("similitud: {}".format(cos_similarity(Naranja, Betabel)))
```

```python
Betabel = np.array([0,0,0.95])
Mujer = np.array([1,0.02,0.01])
Naranja = np.array([0.01,0,0.97])


origin = [0],[0]
plt.quiver(*origin, Naranja,Betabel, color = ['r','g','b'], scale=5)
plt.show()
print("similitud: {}".format(cos_similarity(Naranja, Betabel)))
```

```python
Hombre = np.array([-1,0.01,0.01,0.03])
Mujer = np.array([1,0.02,0.01,0.02])
Rey = np.array([-0.95,0.93,0.02,0.70])
Reyna = np.array([0.95,0.95,0.01,0.69])

print("similitud: {}".format(cos_similarity(Reyna, (Rey-Hombre+Mujer))))
```
